```python
def p2(data):    
    instructions = parse_instructions(data)
    s = ""
    for cycle in range(240):
        x = instructions[cycle]
        if cycle%40 in range(x-1, x+2):
            s += "#"
        else:
            s += "."
        if cycle in range(39, 220, 40):
            s += "\n"
    return s
# print("0"*10, "1"*10, "2"*10, "3"*10, sep="")
# print("".join(map(str, range(10)))*4, sep="")
# print(p2(example_input))
assert (res := p2(example_input)) == """##..##..##..##..##..##..##..##..##..##..
###...###...###...###...###...###...###.
####....####....####....####....####....
#####.....#####.....#####.....#####.....
######......######......######......####
#######.......#######.......#######.....""", f"{res=}"

print(p2(pathlib.Path("day10.txt").read_text()))
```
